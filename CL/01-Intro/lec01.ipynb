### Введение в Python. Характеристики языка. Числовые типы данных. 

На сегодняшнем занятии мы с вами поговорили о том, как вообще устроен мир программирования, что из себя представляет любой язык программирования и как мы взаимодействуем с его помощью с компьютером. 

Язык программирования - это, как и человеческий язык, в первую очередь система; у каждого ЯП есть своя философия и, если угодно, парадигма. Существуют две взаимосвязанные вещи: идея языка, его концепция, грамматика - и имплементация этой идеи на практике, то есть, в случае питона это интерпретатор. 

Python был разработан Гвидо ван Россумом в 1991 году. Основные его характеристики как ЯП:

- это язык высокого уровня (т.е. близко к человеческому)
- это язык интерпретаторного типа 
- это язык с динамической типизацией
- синтаксис питона максимально прост и удобен, даже лаконичен
- питон за вас делает многие вещи, но из-за этого он может работать медленно и ограничивается использованием одного процессорного ядра

Тут следует сказать пару слов о процессорах. 

Процессоры первоначально задумывались людьми именно как устройства для действий над числами, поэтому каждый процессор имеет набор заложенных в него команд (инструкций) для арифметических операций. Способы, которыми инженеры учат процессоры считать, называются микрокодом - то есть, у процессора есть свой (очень ограниченный) набор этих инструкций, которыми можно им управлять. Также у процессора есть кеш (L1-cache, L2-cache): если когда-нибудь будете покупать себе процессор или компьютер, можете посмотреть, размер этого кеша обычно указывается в характеристиках. Например, когда нам нужно сложить два числа, мы должны эти числа положить в память к процессору и отдать инструкцию для сложения. 

Микрокод - это самый низкий уровень программирования, им владеют только те самые волшебные инженеры. Команды микрокода сильно зависят от модели процессора: если у вас Intel i7, набор команд будет один, а если AMD Ryzen 3 - то совершенно другой. 

Напрямую с процессором связывается машинный код, который представляет собой набор инструкций в единицах и ноликах и непосредственно взаимодействует с микрокодом. Когда-то давно люди умели писать программы только на машинном коде (в интернете можно найти фотографии старинных перфокарт, где приходилось этот код еще и набивать на картонках специальным образом). На машинном коде писать было грустно и сложно, а главное, очень тяжело искать ошибки и очень легко их допустить, потому люди придумали сделать набор мнемоник и простенькую програмку, которая бы сопоставляла эти мнемоники с реальными командами машинного кода. Так появился ассемблер (assembly language). Он тоже сильно зависит от версии процессора и очень сложен в освоении (студенты второго курса специальности "программные системы" рыдали в три ручья весь семестр), близок к машинному коду и потому кажется странным для непривычных людей, зато очень быстро работает. 

Можете посмотреть, как выглядит программа на ассемблере, и ужаснуться. Это всего лишь эквивалент нашего print('hello world'). 


             global    _start

              section   .text
    _start:   mov       rax, 1                  ; system call for write
              mov       rdi, 1                  ; file handle 1 is stdout
              mov       rsi, message            ; address of string to output
              mov       rdx, 13                 ; number of bytes
              syscall                           ; invoke operating system to do the write
              mov       rax, 60                 ; system call for exit
              xor       rdi, rdi                ; exit code 0
              syscall                           ; invoke operating system to exit

              section   .data
    message:  db        "Hello, World", 10      ; note the newline at the end

Чуть более высокий уровень - это такой язык, как С. С - знаменитейший язык; ему уже под полтинник, а люди до сих пор пишут на нем код. Компилятор языка С (его переводчик) написан в ассемблере. С уже гораздо больше похож на человеческий язык, но все еще не очень. В С программисту приходится самостоятельно управлять тем, как его программа работает с памятью компьютера, и многими другими ужасными вещами (любой паблик для программистов ломится от мемов про указатели и про то, как никто их не понимает). Когда мы пишем код в С, получается, что код переводится сперва с человеческого языка на С, потом на ассемблер, потом на машинный код и наконец на микрокод. 

Hello world на языке С выглядит уже не столь ужасно:

    #include <stdio.h>
    
    int main()
    {
        puts("Hello world");
    }

Наконец, есть такие языки, как питон. Самая известная и считающаяся стандартной имплементация питона, CPython, написана на языке С. То есть, для того, чтобы перевести компьютеру скрипт на языке питона, нужен "переводчик"-интерпретатор; он и написан на С, вы можете самостоятельно заглянуть в исходники, они лежат в папке include, которая живет в папке с питоном. Там вы найдете файлы с расширением h: это С-шные header'ы. Если вы немного поизучаете С, поймете, что в них написано. 

Имплементации питона могут быть написаны также на других языках или просто по-другому. Наиболее известные имплементации, которые вам могут встретиться:

- Cython
- IPython
- Jython
- PyPy
- Iron Python

В этих версиях может отличаться синтаксис (так, Cython похож на помесь С и обычного питона, а Jython сильно попахивает Java). 

Напоминаю, главное отличие интерпретатора от компилятора заключается в принципе его работы. Компилятор сразу переводит на машинный код всю программу, написанную человеком. Поэтому если в коде какие-то ошибки, компилятор их вываливает во время перевода. Если у вас язык компиляторного типа (С, C++, Java, Pascal, Delphi...), ваша программа проходит через цикл:

- стадия написания кода
- compile-time
- runtime

Соответственно ошибки в ее жизни могут быть двух типов: compile-time error & runtime error. Возможно, вам доводилось видеть сообщения о последних, когда вылетает какая-нибудь программа. 

Преимущества ЯП с компилятором: во-первых, скомпилированный машинный код уже не зависит от языка (собственно, он о языке ничего не помнит, это как китайский текст, переведенный с арабского: исходник ему уже не нужен), во-вторых, он сравнительно мало весит. Недостатки ЯП с компилятором: об ошибке не узнаешь, пока не скомпилируешь, а у языка С все сообщения об ошибках выглядят примерно как "Segmentation fault, вы умрете". 

Преимущества и недостатки ЯП с интерпретатором прямо противоположные. Вы не можете запустить скрипт питона на машине, на которой не установлен его интерпретатор, а если захотите сделать из него exe, придется в этот exe запихнуть весь питон. Зато вы можете экспериментировать с питоном (ошибок compile-time у него не бывает: только runtime), и он очень подробно сообщает, где вы провинились. 

Итак, компилятор и интерпретатор еще можно сравнить с письменным и устным переводчиком (кстати, очень близка англоязычная пара translator - interpreter). Существуют еще такие термины, как Ahead-of-Time (AOT) и Just-in-Time (JIT), которые как раз про то, когда нам вываливаются ошибки: до рабочего запуска программы или прямо во время. Соответственно, питон использует подход JIT. 

Важное отличительное свойство питона также заключается в том, что для того, чтобы выполнять скрипты, он использует Python Virtual Machine: она запускается, когда вы запускаете интерпретатор, ловит все ваши команды и выполняет их. Также PVM делает многие важные вещи за вас: например, ходит за вами с веником и вычищает все ненужные объекты, а также самостоятельно выделяет под них же память, когда нужно (в языке С вам пришлось бы думать о том, сколько вам нужно выделить в оперативной памяти места на хранение каждого числа, строчки и массива, а еще не забывать эту память освобождать, а то другие приложения обидятся). 

Но с этой благодатью питона связаны и некоторые его ограничения, а также неприятные особенности, от которых некоторые программисты С приходят в бешенство. 

Одно из известных ограничений - это [Global Interpreter Lock](https://en.wikipedia.org/wiki/Global_interpreter_lock) (GIL). Что это означает? А чтобы питон со своими автоматизированными процессами сбора мусора и выделения памяти ничего вам не убил, его насильно ограничивают только одним процессорным ядром. Напоминаю, что современные процессоры, как правило, имеют больше одного ядра (средний процессор сегодня имеет 4-8 ядер). Соответственно, даже если вы случайно запустите бесконечный цикл, питон вам компьютер не повесит (а только одно ядро будет страдать). Из минусов - одно ядрышко работает медленно, но мы с вами, если будет время, научимся распараллеливать процессы. 

Ну а неприятная особенность, связанная с автоматическим управлением памятью - это изменяемые и неизменяемые объекты питона, о чем речь пойдет потом и очень много и часто. 

Помимо этого, в характеристиках питона мы записали: "динамическая типизация". А это что такое?

В отличие от человека, компьютер сам по себе не различает числа, буквы, звуки, картинки и так далее: для него это все - цифры (любой объект представляется в виде последовательности цифр, с картинками и текстами в таком виде мы с вами еще поработаем во 2 и 3 семестре). Как же он может различить, когда "3" + "4" должно стать "7", а когда "34"?

Очевидно, что надо ему сообщать прямо, хотим ли мы работать с символом "3" как с числом или как со строкой, например. Типизация - это очень важное понятие в программировании, без него на самом деле не может обойтись ни один язык хоть в каком-нибудь виде. 

Но ЯП реализуют эту типизацию довольно по-разному. Так, например, в языках типа паскаля все переменные до единой, даже жалкие каунтеры, необходимо с помпой объявлять в начале программы с указанием типов, а в процессе программы упаси боже попытаться присвоить в переменную типа int объект типа char! Это называется строгая типизация. 

В питоне типизация динамическая: это означает, что, во-первых, вы переменные объявляете когда и где хотите, во-вторых, вы явным образом не указываете их тип (питон сам угадывает), а в-третьих, в любую переменную можно запихнуть сперва число, а потом строку, а потом что угодно. На самом деле у питона, конечно, все равно все объекты относятся к конкретным типам, более того, он тайком от вас делает неявные преобразования, когда ему это нужно. 

Последнее в философско-исторической части лекции: парадигмы программирования. 

Что это такое? Это про совокупность взглядов о том, как писать код. 

Когда компьютеры только появились, программы писались просто в строчку списком команд, а если нужно было выполнить команду не по порядку, то использовалась команда goto, которая отправляла компьютер на случайную строчку. Такие программы, конечно, были очень запутанными, и если компьютеры еще были в состоянии отследить все goto, то люди не очень, а между прочим умение читать код тоже очень важно: например, нужно читать код коллег, код в каких-нибудь туториалах, а я постоянно ваш код читаю...

Поэтому американский математик и программист Э. Дейкстра написал статью, в которой призывал отказаться от оператора goto и стараться свои программы четко структурировать. Так появилась парадигма структурного программирования. В основном это про то, чтобы стараться программы писать по четким блок-схемам. 

Но программы делались все сложнее, а еще часто приходилось один и тот же набор команд повторять в самых разных местах, и тогда люди придумали функции: почему бы не взять набор постоянно повторяющихся друг за другом команд и не выделить их в отдельную вещь, чтобы сразу скопом вызывать в нужном месте? Так появилась парадигма функционального программирования (оно еще тесно связано с математикой). Мы весь мир представляем как набор действий (функций), а объекты нам не важны, мы по идее можем любой объект в функцию запихнуть. 

Самая сложная для понимания парадигма - это объектно-ориентированное программирование. Это когда у нас, наоборот, в центре всего внимания - объекты, иерархии объектов, а все остальное - только их свойства (атрибуты). 

Подробнее о парадигмах будем говорить потом; поначалу мы будем работать только в рамках структурной парадигмы, потом плавно перейдем к функциональному программированию и ООП. 


Питон называют мультипарадигмальным языком: это означает, что вы можете писать в нем код в любой парадигме. Не все языки такие гибкие: например, в Java вы можете писать только объектно-ориентированный код, а в С вряд ли получится в чистое структурное программирование или в ООП.

#### Числа в питоне

Ну что же, вот типы чисел, которые знает питон (они называются встроенными типами вместе с некоторыми другими):

- int: целые числа
- float: числа с плавающей точкой
- complex: комплексные (иррациональные) числа
- bool: подвид int, всего два объекта такого типа - False (0) и True (1)

Любой объект можно (попытаться) преобразовать в любой из этих типов с помощью одноименных функций. 

- Если мы преобразуем строку, то питон пытается перевести ее в число. Если в строке есть лишние символы, вывалится ошибка. 
- Если мы преобразуем float в int, то дробная часть просто отсечется. 
- Если мы преобразуем в bool что угодно, то что угодно превращается в True, если оно не пустое / не равно нулю. 

Обратите внимание, что функция input() возвращает **всегда** строку. То есть, когда вы пишете:

```python
a = input()
```

То в переменной а всегда будет строка, что бы вы ни ввели. Чтобы превратить то, что в переменной а, в число, нужно применить к нему функцию int:

```python
a = int(input())
```

Заметьте: функция input() оказалась вложена в функцию int(). Это то, что называется pythonic way: питонисты любят вкладывать выражения одно в другое и не создавать лишние переменные (однако всему есть свой предел). 

С любыми практически числами питон умеет производить арифметические операции:

    + - сложение
    - - вычитание
    * - умножение
    / - деление (точное; всегда возвращает float)
    // - целочисленное деление (возвращает только целую часть)
    % - остаток от деления (например, 5 % 3 = 1, потому что в 5 помещается только одна тройка)
    ** - возведение в степень

Очередность этих операций при выполнении такая же, как и в обычной школьной математике, и точно так же можно использовать скобки:

```python
hypo = (a ** 2 + b ** 2) ** 0.5
```

Каждая из этих операций принимает два операнда и возвращает новый объект. Все операции возвращают объект того типа, какого они получили, кроме деления /, которое всегда возвращает float. Если мы в качестве операндов другим операциям передаем одно число float, а другое int, то вернется float, как более точное (потому что мы int можем перевести без потерь во float, а наоборот теряем дробную часть: это пример неявного преобразования типов в питоне). 

Числа в питоне - это **неизменяемые** объекты. Их еще называют hashable (в современных версиях, по-моему, начиная где-то с 3.7, так их и называют). Это что значит? 

Хеш - это способ быстро найти какой-нибудь объект. Представим себе, что память компьютера - это шкаф с кучей ящиков, по которым разложены предметы. Мы можем перерывать все ящики по очереди, чтобы найти вон тот зеленый теннисный мячик, а можем каждому объекту приписать кодовое число таким образом, чтобы объекты с числами от 1 до 10 лежали в первом ящике, от 11 до 20 во втором и так далее: тогда мы мгновенно найдем, какой ящик нам нужен. 

Хеш примерно так и работает. Сам по себе хеш - это какое-то короткое число, ключ, который приписывается конкретному объекту. Каким образом он приписывается? Для этого обычно используются хеш-функции. Как они устроены, знать не обязательно, но как минимум на англоязычной википедии про это точно все есть. 

Соответственно, всем объектам, которые он складывает в память, питон приписывает хеши... или не всем. Ну вот как раз тем, которые hashable, приписывает, и благодаря этому может почти мгновенно их находить, и это же означает, что объект, который имеет хеш - какой-то цельный предмет, который лежит в памяти в конкретном месте. Поэтому его еще и можно назвать неизменяемым: в том плане, что теннисный шарик мы частично заменить на что-нибудь другое не можем, только выкинуть и взять новый. Другое дело шахматы, когда мы можем из набора взять пешку и заменить ее на пешку из другого рабора - но об этом поговорим чуть позже. 

Так вот, что следует из того, что числа неизменяемые?

Когда мы делаем вот такое:

```python
a = 1
a = a + 1
```

Мы на самом деле (в глазах питона) не увеличиваем число 1 на единицу, а заменяем единицу, которая хранится в переменной а, на двойку. 

Чуть подробнее: 

В нашей памяти где-то лежит объект 1. У него есть свой хеш. Когда мы присваиваем его в переменную а, переменная связывается с этим объектом и хранит в себе его адрес. Когда мы хотим увеличить единичку на 1, в момент, когда мы велим компьютеру сделать a + 1, он берет единичку из а и еще одну другую единичку, складывает их в кеш к процессору, заставляет процессор произвести операцию сложения; процессор возвращает в качестве ответа "2": новый объект. Этот объект теперь тоже лежит у нас в памяти где-то. Дальше мы присваиваем его в переменную а, и теперь а теряет свой поводок от первой единички и начинает указывать на двоечку. А что происходит с единичкой? За ней приходит питоновский сборщик мусора и скушивает ее. 

(На самом деле не совсем так: при запуске PVM для оптимизации скорости обычно заранее складывает в память первые несколько десятков целых чисел, поэтому единичка все равно где-то в памяти болтаться будет, а вот большое число действительно исчезнет). 

Соответственно, еще немного про адреса объектов в памяти:

функция id() позволяет нам получить уникальный идентификатор объекта, его "телефонный номер", по которому питон его вызывает.

#### Условные выражения, операторы, инструкции

Тип bool мы с вами уже знаем. В питоне также есть условные операторы:

    < - меньше
    > - больше
    <= - меньше или равно
    >= - больше или равно
    == - равно
    != - не равно

Запомните, что выражение с условными операторами **вычисляются** так же, как арифметические, и возвращают объект типа bool.

```python
a = 1 < 2 
type(a)
```

И постарайтесь не путать оператор присваивания = и оператор сравнения ==. 

К условным операторам также относятся слова and, or, not. Они обозначают соответственно логическую дизъюнкцию, конъюнкцию и отрицание. Сами по себе они возвращают обычно тип bool:

```python
print(1 < 2 and 2 < 3)
print(1 < 2 or 1 > 2)
print(not 1 == 0)
```

Но есть нюанс.....

```python
1 or 5 > 3
```

```python
4 and 2 < 3
```

Чем это вызвано? 

Питонической ленью!

Оператор or принимает значение "истинно" в случае, если хотя бы одна из его частей истинна; поэтому он тут же кидается проверять первый встретившийся объект, и если этот объект истинный, то он даже не смотрит все остальное (это удобно в случаях, когда вы, например, хотите проверить что-то вроде if a != None and a > 3 -- если бы а было равно None и питон попытался сравнить его с тройкой, вывалилась бы ошибка, а так прокатит). Также or возвращает тот самый истинный объект в неизменном виде (ну что ему стоит взять и единицу вернуть as is?). 

Оператор and принимает значение "истинно" только в том случае, если все его части истинны. Поэтому ему приходится проверять все-все, и он возвращает результат проверки всего, так что его результат - это всегда какой-то bool. 

Такая особенность or еще иногда используется для того, чтобы избежать деления на ноль: 10 / (int(input) or 1)

Наконец, у нас с вами есть условные инструкции. В рамках структурной парадигмы это как раз те самые ромбики в блок-схеме, где нужно пойти налево или направо. Как они устроены в питоне:

```python
a = int(input())
if a < 0:
    print('Число отрицательное')
elif a > 0:
    print('Число положительное')
else:
    print('Вы ввели ноль')
```

Что здесь что?

**if** - это оператор. Сразу за ним должен следовать какой-нибудь (любой!) объект. Это может быть объект типа bool (помним, что выражение a < 0 вычисляется и возвращает как раз такой объект), а может быть любой другой объект, который if обязательно неявно превратит в bool. После этого объекта стоит двоеточие: оно означает, что начинается **тело** инструкции if. 

Тело пустым быть не может: вы обязаны написать хотя бы одну малюсенькую команду, чего компьютеру делать, если ваше условие выполнилось (а иначе зачем условие было писать?). В примере выше в теле оператора if находится функция print(). Чтобы показать питону, что это именно тело, его отделяют отступом: четырьмя пробелами (в современных средах разработки табуляция автоматически заменяется именно на 4 пробела). Если вы забудете про отступ, питон заругается на ошибку синтаксиса. Это одна из особенностей питона: в других языках обычно тело конструкции, функции и т.п. выделяют, например, фигурными скобками или словами BEGIN & END (можете еще раз взглянуть на пример кода на С), и хотя там тоже принято делать отступы из четырех пробелов, там они совсем не обязательные. В питоне - **обязательные**. 

Вот все остальные части после этого - факультативные, вы можете пропустить любую из них. elif = else if = такой блок вообще может повторяться сколько угодно, например:

```python
if a == 1:
    <do something>  # это, конечно, не команда, я для наглядности написала
elif a == 2:
    <do something>
elif a == 3:
    ...
```

else тоже может быть, а может не быть; else выполняется "во всех остальных случаях, которые не подошли ни в один if-elif). Если же они есть, то if и относящиеся к нему elif-else должны быть вертикально на одном уровне. 

Условные инструкции могут быть и вложенными:

```python
if a < 1:
    if a > 0:
        print(a)
    else:
        print()
else:
    print('big a')
```

В таком случае приходится внимательно следить за лесенкой. 

В теле инструкции, конечно, может быть больше одной команды. 

Последний небольшой оффтопик, о котором мы успели поговорить - это форматирование строк. 

Вообще в питоне есть три способа форматирования, мы пока обсудили только один, и то лишь в общих чертах: это f-строки. 

f-строка - это такой шаблон, в который питон за вас в нужные места подставляет значения переменных. 

```python
a = 1
b = 2
print(f'a is {a}, b is {b}')
```

В этом примере текст "a is ..., b is ..." - неизменяемая часть шаблона; он будет напечатан принтом по-любому. А вот фигурные скобки говорят питону, что в эти места нужно подставить изменяемые вещи (это еще называется джокеры), в нашем примере - значения переменных а и b. В фигурную скобку можно поместить и выражение (а + b), и даже функцию. Главное, чтобы там не было кавычек, а то питон не поймет, где у него граница строки. 

Также у функции print() есть свои необязательные параметры, которыми можно пользоваться:

```python
print(a, b, sep=':', end='!')
```

#### Last, but not least: конвенции

Конвенции - это правила "орфографии", которые соблюдать для работоспособности кода необязательно, но из вежливости к другим людям, которые ваш код будут читать, очень желательно. 

Как правило, если ваш код написан без соблюдения конвенций, у людей, читающих его, сразу возникает ощущение, что вы программировать не умеете. Это может в будущем сказаться на собеседованиях и проч. 

Конвенции для питона разрабатывал его автор ван Россум с коллегами; оригинал можно почитать [здесь](https://peps.python.org/pep-0008/). Самые базовые правила, которые вам следует соблюдать, чтобы я на вас не бухтела:

- Любой оператор (арифметический, условный) должен быть окружен пробелами: a + b, c > d, a = 1
- После запятой ставьте пробел. 
- между функцией и ее скобками пробела быть не должно: print(), а не print ()
- = не окружается пробелами, если оно находится в именованных параметрах функции: print(... sep='\t')
- старайтесь называть переменные понятными именами: если у вас переменная-счетчик, назовите ее count, counter или на худой конец c. 
- не называйте переменные с заглавной буквы: есть некоторые переменные, которые положено называть с заглавной буквы, но мы до них еще не дошли. 

Особенно я почему-то бешусь на пробелы, имейте в виду. Первые пару домашек без конвенций я еще прощу (но буду бухтеть), а потом при грубых нарушениях конвенций я буду **отказываться их проверять, пока не поправите**. 

#### Строки и методы строк

Первое, что нужно знать про строки - это как они в питоне обозначаются. Мы уже выяснили, что целые числа просто пишутся как числа, числа с плавающей точкой пишутся с точкой (не с запятой!), а строки берутся в кавычки. Существует несколько вариантов, как можно закавычивать строки:

    'Hello'
    "Hello"
    
    '''Hello'''
    """Hello"""
    
Теоретически любой вариант сообщает питону, что перед ним строка. Какие между этими способами различия?

- С точки зрения интерпретатора: тройные кавычки позволяют строку разрывать переносами, а одинарные (в любом варианте) требуют, чтобы мы писали \n:

        'Hello\nworld'
        """Hello
        world"""
        
- С точки зрения конвенций: 1) двойные кавычки " чаще используются со строками, если вообще используются. А еще можно их использовать, чтобы внутри строки апострофы не нужно было экранировать: "King's crown" vs 'King\'s crown' 2) тройные кавычки используются для того, чтобы просто писать длинные комментарии, а тройные двойные кавычки ("""""") используются для того, чтобы писать docstrings. Что это, обсудим на следующей паре. 

Итак, строка - это **неизменяемый итерируемый** тип. Со строками менее очевидно, но:

```python
a = 'hello'
a += ' world'
a
```

Объект 'hello' не изменился: мы просто в переменную а сложили другой объект. 

Итерируемость строк в первую очередь выражается в индексировании. Строка состоит из отдельных символов, к которым можно обращаться по их порядковым номерам:

```python
a = 'qwerty'
print(a[0], a[-1], a[1:3], a[1:4:2])
```

В квадратных скобках можно указывать разные вещи: если указать одиночное число, то питон обратится к конкретному символу под этим номером, а если указать два числа через :, то питон это воспримет как *срез* - кусочек строки (подстроку). Срезы обычно легко запоминать так:

    [start:stop:step]
    
Причем любые две части можно опустить. Если не указать start, то питон по умолчанию возьмет срез с нулевого элемента строки. Если не указать stop, то по умолчанию срез дойдет до последнего символа в строке. step по умолчанию равен единице. 

Можно указывать отрицательные числа: -1 означает последний символ в строке, сколько бы их там ни было, -2 - предпоследний и т.д. Шаг -1 означает, что мы пойдем по строке задом наперед. 

Теперь пару слов о функциях и методах. Мы с вами уже довольно знаем функций в питоне:

    print()
    input()
    int()
    str()
    float()
    bool()
    type()
    
Функции могут что-то *возвращать*, а могут ничего не возвращать. Сравните:

```python
returns = input()
doesnt_return = print()
print(f'input возвращает: {returns}, print не возвращает: {doesnt_return}')
```

А еще есть методы. Например, методы есть у строк. Они выглядят так:

    a.strip()  # возвращает новую строку без пробельных символов по краям
    
Методы - это такие же функции, но они прикреплены к конкретным типам данных. Например, у строк есть свои методы, которые нельзя применять к числам. Часто методы встроенных типов данных называются одинаково - но *по-разному реализованы внутри*. Метод пишется через точку от объекта, к которому применяется: точка в синтаксисе питона в общем случае означает принадлежность. 

Методы точно так же, как и функции, могут что-то возвращать, а могут ничего не возвращать. Методы строк всегда что-нибудь возвращают, потому что строки неизменяемые. 

Какие мы с вами рассмотрели:

- Удаляют лишние пробельные символы (возвращают строки):

        a.strip() # удалит все пробельные символы слева и справа. Можно в скобках передать свой символ или набор символов, он их все отрежет
        a.rstrip()
        a.lstrip() # соответственно удаляют только слева или справа
        
- Работают с индексированием, возвращают числа:

        a.find('x') # ищет х в строке, возвращает индекс первого слева, если не нашел - вернет -1
        a.rfind('x') # то же, но справа
        a.index('x') # вернет индекс х
        a.count('x') # подсчитает, сколько в строке встретилось х
        
- Работают с регистром, возвращают строки:

        a.upper() # ВЕРХНИЙ
        a.lower() # нижний
        a.capitalize() # Одно слово с заглавной
        a.title() # Каждое Слово С Заглавной
        
- Проверяют содержимое строки, возвращают bool (True/False):

        a.startswith('x')
        a.endswith('x')
        a.isdigit()
        a.isalpha()
        a.isalnum()
        a.isspace()
        a.istitle()
        a.islower()
        a.isupper()
        
Не метод, но логический оператор: проверка членства

    'x' in a  # in - оператор; такое выражение вернет True, если подстрока 'x' содержится в а. 
    
- Заменяет символы, возвращает строку:

        a.replace('что меняем', 'на что меняем')
        
- Возвращает список:

        a.split()  # бьет по пробелу на подстроки
        
- Склеивает элементы итерируемого объекта, перемежая их строкой:

        a.join(iterable)

```python
' '.join('abcde')
```

А можно так:

```python
' '.join(input().split())
```

Тут что происходит? Сперва питон выполняет input(), считывает мое 'мама мыла раму', потом разделяет на подстроки 'мама', 'мыла', 'раму'. Это список: тоже итерируемый объект. Список потом склеивается с пробелом, к которому мы применяем метод join, и получается обратно наша строка. 

#### Кодировки

Буквы и вообще любые символы в компьютере представляются в виде чисел. Давно люди решили на кодировку букв человеческого алфавита выделить по байту на символ. Байт - единица информации, содержит в себе 8 битов. Бит - это минимальная единица информации, он может принимать значения либо 0, либо 1. Соответственно, если кодировать символы одним байтом, то получается, что один символ будет выглядеть как-нибудь так:

    11011011
    
Сколько можно закодировать так символов? Бит может принимать только два значения, а в байте их 8 штук. Значит, уникальных номеров будет:

```python
2 ** 8
```

Маловато! В эти 256 символов влезают стандартные знаки пунктуации, цифры, латинский алфавит и еще немножко места остается для, например, кириллицы. 

Такая кодировка - однобайтовая - называется в общем смысле ASCII. Почитать про нее подробнее можно на Википедии. 128 символов в ней отведены под латинский алфавит, а остальные 128 могут занимать символы другого языка. Соответственно, в каждой почти стране кодировка своя, например, для русского языка есть варианты:

    cp1251
    cp1252
    koi-8
    ...
    
Их еще редко-редко можно увидеть в почте, когда скачиваете архивом приложения, почтовик может спросить, в какой кодировке качать. Также операционная система Windows (по крайней мере, до 10 версии включительно) работает в кодировке cp1251. Если вам интересно, кодировку на вашей системе можно проверить этим кодом:

```python
import locale

locale.getpreferredencoding()
```

У меня utf-8, потому что у меня Линукс. 

Что же это за utf-8? 

Ну, очевидно, что с 256 символами было жить грустно (а как же китайцы?), поэтому люди решили не жмотиться и отвести под один символ сразу 4 байта. 4 байта = 32 бита, итого:

```python
2 ** 32
```

Вот этого уже на все алфавиты, включая марсианские, хватит! 

Общее название для четырехбайтовых кодировок - юникод (Unicode). Питон работает в юникоде! И поэтому поддерживает любые алфавиты (в отличие от, например, Java, в котором с ними довольно тяжело бороться, если ты живешь на Windows...). И даже эмодзи. 

У четырехбайтовых кодировок, однако, есть еще разновидности. В частности, кодировать четырьмя байтами латинский алфавит и супер-частотные пробелы и точки невыгодно: ведь размер текста будет увеличиваться в 4 раза! А занимать это место будут бесполезные ноли (первые три байта будут пустые потому что). 

Выход - самые частотные символы кодировать одним байтом, например, менее частотные - двумя, а редкие - четырьмя. Это кодировка utf-8. 

Другой вариант - это когда мы используем только 2 и 4 байта для кодирования и получаем utf-16. 

Ну а самый жирный и менее всего используемый - utf32. (Их можно писать с дефисом и без, если что). 

Так вот, и питон умеет превращать нам числа в буквы (считая, что число - это порядковый номер символа в юникоде) и наоборот. 

```python
print(ord('a'))
print(chr(97))
```

Соответственно, функция ord() находит порядковый номер (в десятичной системе) для символа, а функция chr превращает номер в символ. 

#### Форматирование строк

Форматировать строки можно двумя (даже тремя, если считать устаревший способ форматирования питона 2) способами:

- f-strings: f'This is variable {a}'
- метод строк format: 'This is variable {}'.format(a)

Идейно они практически ничем не различаются. В фигурные скобки мы можем подставлять любые переменные, в f-строках явным образом прописывая их имена, а в методе format передавая аргументами. В format можно их еще нумеровать:

```python
a = 1
b = 2
c = 3
'c: {2}, b: {1}, a: {0}'.format(a, b, c)
```

Номера указывают, в каком порядке подставлять, а если не указать, то автоматически все подставится слева направо в том порядке, в каком указали. 

Форматированные строки на самом деле много чего клевого умеют. 

Во-первых, выравнивание текста:

```python
f'{a:*<4}|{b:*>4}'
```

Что здесь что? Двоеточие внутри фигурных скобочек сообщает питону, что мы хотим задать какие-то параметры. Первый из символов - это fill: то, чем мы заполняем пространство, отведенное под эту переменную (я указала звездочку). Второй символ - собственно выравнивание (слева, справа, по центру). Последняя циферка - это сколько символов под это поле отводить. Посмотрим, как это будет работать в цикле:

```python
for i in range(1, 100, 10):
    print(f'{i:*^7}')
```

Также можно, например, округлять числа прямо в форматированной строке:

```python
from math import pi  # из библиотеки для математиков импортирую число пи

print(f'pi is {pi:.2f}')
```

здесь .2 означает, сколько знаков после запятой оставлять, а f означает, что мы хотим формат float (эти форматы разные бывают, попробуйте убрать букву f и посмотреть, что получится). 

Все те же самые штуки работают и с методом format, только тогда может быть так:

    '{:*<4}|{:*>4}'.format(a, b)
    
А еще можно задавать размер окна с помощью переменной:

```python
print('{:*<{width}}|{:*>{width}}'.format(a, b, width=input()))  # да, так тоже можно! 
```

## Циклы

На первом, по-моему, занятии мы с вами немного говорили про парадигмы программирования. Напомню самую простую, в которой мы с вами пока что и работаем: структурную. 

Структурное программирование - это когда мы свои программы разрабатываем сперва в виде алгоритмов (блок-схем), а потом уже пишем код. Что там в этих блок-схемах бывает? ромбики-ветвления и циклы. Ну например:

![блок-схема](block.png)

Здесь у нас есть и ветвление - если дошли до края или если не дошли - и цикл: пока не выполнится условие "дошли до края", будем кружиться и шагать. 

В питоне есть средства для реализации таких вот кружений: циклы. Их существует две разновидности. 

1. Цикл, который выполняется, пока выполняется какое-то условие: while
2. Цикл, который выполняется заранее известное количество раз: for

Итак, как устроен while?

```python
a = input()       # считываем ввод пользователя
while a:          # если пользователь просто нажмет Enter, получится пустая строка в а, а bool('') - False. 
    print(a)      # выводим эхом в печать
    a = input()   # считываем следующее а. Если этой команды не будет, цикл будет бесконечным! Ведь а не меняется, а нужно, чтобы менялось
```

То есть, while после себя ожидает какой-то объект типа bool. Туда можно написать в принципе все, что угодно, но это все питон попытается вычислить и превратить в такой объект. Как и в if, удобно писать в while условные выражения (a < 0) или методы, которые возвращают объект bool (посмотрите, какие мы уже знаем для строк). 

Цикл for устроен чуточку сложнее:

```python
for char in 'abcdef':   # мы можем передать любой итерируемый объект, будь то литерал или переменная
    print(char)
```

У цикла for есть *переменная цикла*: в примере это char. Что реально происходит на каждой итерации? Вот у нас есть строка из 6 символов. На первой итерации питон берет первый символ в строке и делает char = 'a'. Потом он делает все, что написано в теле цикла, и возвращается снова к началу. Теперь то же самое ему нужно проделать со вторым символом строки, и он делает char = 'b'. Таким образом, **переменная цикла меняется на каждой итерации**. Она не перестает существовать после выполнения цикла:

```python
char
```

Но использовать ее вне цикла - дурной тон, потому что обычно вы заранее не знаете, что в ней может оказаться. 

С помощью цикла for можно еще итерироваться по индексам строк, а не напрямую по их символам:

```python
a = 'qwerty'
for i in range(len(a)):
    print(a[i])
```

Для строк не имеет большой разницы, как именно мы итерируемся, а вот для список имеет и значительную, до этого еще дойдем. Обратите внимание здесь на две вещи:

Функция len() возвращает длину итерируемого объекта. У строк есть длина - количество их символов. Не заводите переменные с именем len, пожалуйста! Вы тогда перезапишете эту функцию и не сможете ей воспользоваться, а она одна из самых полезных. 

Функция range() работает примерно как срезы (range(start, stop, step)) и возвращает специальный объект типа range (диапазон). По нему можно итерироваться, например, ну и конструкция типа приведенной выше - супер-распространенная, она должна от зубов отлетать. 

У циклов есть и некоторые (необязательные) продолжения. Во-первых, есть два оператора, которые используются с циклами и if-else statements: break & continue. Что они делают?

```python
a = input()
for char in a:
    if char == 'x':
        break
    print(char)
```

break, как мы видим, - аварийная остановка цикла, она мгновенно выкидывает вас из цикла, игнорируя все, что после нее. Именно поэтому никакой else после break не нужен - он избыточный: если выполнится if, то нас просто выкинет и все остальное пофиг. 

```python
a = input()
for char in a:
    if char == 'x':
        continue
    print(char)
```

continue полностью из цикла не выкидывает, но игнорирует все команды, которые идут после него, и переводит сразу на следующую итерацию. 

Распространена следующая конструкция с while & break:

```python
while True:
    a = input()
    if a == 'x':
        break
    print(a, end=' ')
```

Таким образом можно избежать повторения команды input(), но условие с брейком рекомендуется писать как можно повыше. 

Еще у обоих этих циклов может быть else: он выполняется только один раз после того, как отработает весь цикл. Если мы совершим аварийную остановку с помощью break, else не выполнится, поэтому можно так проверять, не было ли брейков. 

```python
for char in input():  # да, и так тоже можно, это питон
    if char == 'x':
        break
else:
    print('No breaks!')
```

#### Исключения

Когда мы в коде делаем ошибки (или просто возникает какая-нибудь ошибка во время выполнения нашей программы - не обязательно потому, что мы такие дураки), питон выбрасывает исключения: те самые штуки, которые он выделяет багровым в выводе. 

```python
a = int(input())
```

ValueError - это название исключения. Исключение - тоже своего рода break, только для всей программы разом, когда оно возникает, питон завершает программу и выводит его. В питоне есть встроенные типы исключений, многие из которых скоро намозолят вам глаза: SyntaxError, ValueError, IndexError и другие. Список есть в документации (погуглите exceptions). Эти исключения можно перехватывать таким образом, чтобы программа не вылетала вся сразу. Для этого используется конструкция try-except.

```python
try:
    a = int(input())
except ValueError:
    print('a is not a number')
```

try & except - обязательные части этой конструкции. После try просто ставим двоеточие и в теле пишем то, что пытаемся выполнить, а после except обычно пишется название исключения (исключения - это свои *объекты* в питоне, он их знает и поэтому не нужно брать их в кавычки) и какие-то действия, которые мы делаем, если исключение возникло. 

Есть и две необязательных части:

```python
string = input()
try:
    print(string[4])
except IndexError:
    print('String too short')
else:
    print('String long enough')
finally:
    print('Script ended')
```

Что здесь и когда будет выполняться:

- try - всему голова, поэтому всегда
- except - если возникнет исключение
- else - если исключения не возникнет
- finally - в любом случае выполнится

Последние две вещи могут быть, а могут не быть, или может быть только какая-то одна из них. 

Про исключения немного подробнее поговорим в 3 семестре. 

#### Сложность алгоритмов и Big O

Теперь, когда мы умеем в циклы, рассмотрим такую штуку:

```python
from time import sleep  # импортирую функцию, которая заставляет компьютер ждать и ничего не делать, из библиотеки time

D = 10

for i in range(D):
    sleep(1)
print('Zzzz...')
```

В этом цикле есть некая переменная D, будем считать, что это объем обрабатываемых нами данных (мы ведь и правда проходимся по последовательности от 0 до 9 и на каждом шаге чего-то делаем). Наша программа работает ~ 10 секунд: по одной секунде на функцию sleep(1) и еще какая-то крошечная доля секунды на print(). Что будет, если мы в переменной D укажем не 10, а 100? Программа будет работать примерно 100 секунд. Соответственно, если мы хотим вычислить время работы нашей программы, нам пригодится примерно такая формула:

    Time = D * k + b 
    
где Time - время, D - размер обрабатываемой последовательности, k - время, которое уходит на обработку команды sleep(), а b - время, которое уходит на выполнение функции print() и оно не изменяется, ведь print(), сколько бы у нас ни было D, выполняется только один раз. 

Узнаем? Это уравнение прямой. То есть, время работы нашего алгоритма *линейно зависит* от объема данных. Про такой алгоритм говорят, что у него линейная сложность или что он выполняется за линейное время. У программистов есть для этого своя нотация:  Big O, и они пишут: O(n), то есть, для n данных программа будет работать k * n + b времени. 

Сложность алгоритма бывает и константная:

```python
a = int(input())
b = int(input())
print(a + b)
```

Какие бы числа мы ни ввели, программа будет всегда отрабатывать примерно одинаковое количество времени. В Big O нотации пишут O(c). 

Ну и бывает сложность выше линейной, например, квадратичная:

```python
D = 5

for i in range(D):
    for j in range(D):
        sleep(1)
```

Для D = 5 мы поспим 25 секунд, для D = 10 - 100 секунд и так далее. То есть, у нас $O(n^2)$. 

Конечно, этим виды сложности не ограничиваются, бывает, например, $O(log_n)$, $O(e^n)$ и другие. Логарифмическая - это получше, чем квадратичная, но похуже, чем линейная, например, а экспоненциальная самая ужасная, никогда так не делайте. В идеале нужно стремиться к тому, чтобы код работал с как можно меньшей сложностью (то есть, как можно быстрее). 

Питон, напомню вам, вообще не очень быстрый язык: это обусловлено его синтаксисом и высокоуровневостью. Самые быстрые языки - это ассемблер (супер-быстро), С тоже довольно быстрый, но они сложные для освоения. Тем важнее писать программы в питоне так, чтобы они не работали миллионы времени, если вы можете как-то упростить свой код, упрощайте его. 

Еще есть известный поиск баланса между "быстро - занимает много памяти": обычно вещи, которые работают быстрее, кушают больше оперативы, это очень наглядно будет видно, когда будем изучать нейронки (чем больше батчи, тем быстрее она учится, но можно случайно откушать столько памяти на видеокарте, что перезагрузится компьютер), но оператива - ограниченный ресурс.
